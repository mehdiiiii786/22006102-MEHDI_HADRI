```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

QUESTION **1.1**

```python
moyenneA = np.mean(rendements_A)
stdA = np.std(rendements_A)
medA = np.median(rendements_A)
rendannA = ((1+(moyenneA/100))**12-1)*100
volA = stdA * np.sqrt(12)

```

```python
moyenneB = np.mean(rendements_B)
stdB = np.std(rendements_B)
medB = np.median(rendements_B)
rendannB = ((1+(moyenneB/100))**12-1)*100
volB = stdA * np.sqrt(12)
```
